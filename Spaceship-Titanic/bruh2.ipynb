# Train - CV - Test Split

```python
X_test = df_test
```

```python
y = df_train['Transported']
X = df_train.loc[:,'CryoSleep':'Destination_3']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_cv, y_train, y_cv = train_test_split(X, y, test_size = 0.30, shuffle = True)
```

```python
X_train.reset_index(inplace = True, drop = True)
X_cv.reset_index(inplace = True, drop = True)
y_train.reset_index(inplace = True, drop = True)
y_cv.reset_index(inplace = True, drop = True)
```

# Model Building

```python
import xgboost as xgb
```

```python
dtrain = xgb.DMatrix(data = X_train, label = y_train)
dcv = xgb.DMatrix(data = X_cv, label = y_cv)
```

# Cross-Validation

```python
from xgboost import cv

param = {'num_parallel_tree' : 4, 'max_depth' : 10, 'min_child_weight': 7,
         'objective' : 'binary:logistic', 'eval_metric':'auc'}

xgb_cv = cv(dtrain=dcv, params=param_best, nfold=4,
                    num_boost_round=50, early_stopping_rounds=10, metrics="auc", as_pandas=True, seed=42)

xgb_cv.tail()
```
